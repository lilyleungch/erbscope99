# this is the dataframe with first 5 column local consumption of electricity in Hong Kong where the last 4 columns are local consumption of gas in Hong Kong

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
#temperature =pd.read_csv('CLMMAXT_HKO_.csv')
#csv from https://www.censtatd.gov.hk/en/web_table.html?id=915-91201&full_series=1&download_csv=1
electric= pd.read_csv('electric_gas.csv')
electric.head(3)
electric2=electric.set_index(['Unnamed: 0','Unnamed: 1'])
electric2.plot()
electric2.corr(numeric_only = True)
```

```python
sns.heatmap(electric2.corr( numeric_only = True), annot = True, cmap = 'RdBu_
            r', vmin = -1, vmax = 1, center = 0)
```

```python
electric.head(3)
electric['Combined'] = electric['Unnamed: 0'].astype(str) +'-' + electric['Unnamed: 1'] 
electric.index = electric['Combined']
electric.index=pd.to_datetime(electric.index, format = '%Y-%b')
electric
```

```python
import numpy as np
weather = pd.read_csv('CLMTEMP_HKO_.csv')
#csv from https://data.gov.hk/tc-data/dataset/hk-hko-rss-daily-temperature-info-hko/resource/7f42664a-04ec-4bb4-ac35-d257a554c661
#weather2=weather.groupby(["/Month"])['/Value'].mean()
weather.dropna(inplace=True)
#weather.set_index(['/Year','/Month'])
#weather2=weather.groupby(['/Year', '/Month'])["/Value"].agg(['max','min'])
weather1=weather[weather['/Year']>1978]
weather2=weather1[['/Year', '/Month','/Day',"/Value"]].astype(dtype=np.float32)
weather3=weather2.groupby(['/Year', '/Month'])["/Value"].agg(['max', 'min', 'mean'])
weather3.plot()
```

```python
# change the index into datetime
weather.head(3)
weather['Combined'] = (weather['/Day'].astype(str)) + '-'+ (weather['/Month'].astype(str)) + '-' +(weather['/Year'].astype(str))
weather.set_index(weather['Combined'])
weather.index = pd.to_datetime(weather['Combined'], format = '%d-%m-%Y')
weather

```

```python
#plot starting from 2020 to now

electric3=electric[electric.index.year>2020]
weather4 = weather[weather.index.year>2020]
fig, ax_list = plt.subplots(3,1,figsize=(12, 12))
ax_up = ax_list[0]
ax_down = ax_list[2]
ax_middle = ax_list[1]
ax_up.set_title('electricity usage by all groups')
ax_up.plot(electric3.index, electric3['All groups'], lw= 10)
ax_up.set_ylabel('Terajoule')
ax_middle.set_title('temperature in HK observatory')
ax_middle.scatter(weather4.index, weather4['/Value'].astype(np.float32))
ax_middle.set_ylabel('Celsius')
ax_middle.set_yticks([10,20,30])
ax_down.plot(electric3.index, electric3['All groups.1'],lw = 10)
ax_down.set_title('gas usage by all groups')
ax_down.set_xlabel('Year')
ax_down.set_ylabel('Terajoule')
plt.show()
```

```python
#plot starting from 2020 to now

electric3=electric[electric.index.year>2020]
weather4 = weather[weather.index.year>2020]
fig, ax_list = plt.subplots(3,1,figsize=(12, 14))
ax_up = ax_list[0]
ax_down = ax_list[2]
ax_middle = ax_list[1]
ax_up.set_title('electricity usage by domestic')
ax_up.plot(electric3.index, electric3['Domestic'], lw= 10)
ax_up.set_ylabel('Terajoule')
ax_middle.set_title('temperature in HK observatory')
ax_middle.scatter(weather4.index, weather4['/Value'].astype(np.float32))
ax_middle.set_ylabel('Celsius')
ax_middle.set_yticks([10,20,30])
ax_down.plot(electric3.index, electric3['Domestic.1'],lw = 10)
ax_down.set_title('gas usage by all domestic')
ax_down.set_xlabel('Year')
ax_down.set_ylabel('Terajoule')
plt.show()
```
