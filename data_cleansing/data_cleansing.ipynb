# this is the dataframe with first 5 column local consumption of electricity in Hong Kong where the last 4 columns are local consumption of gas in Hong Kong

```python
import pandas as pd
import matplotlib.pyplot as plt
#temperature =pd.read_csv('CLMMAXT_HKO_.csv')
#csv from https://www.censtatd.gov.hk/en/web_table.html?id=915-91201&full_series=1&download_csv=1
electric= pd.read_csv('electric_gas.csv')
electric.head(3)
electric2=electric.set_index(['Unnamed: 0','Unnamed: 1'])
electric2.plot()
```

```python
electric.head(3)
electric['Combined'] = electric['Unnamed: 0'].astype(str) +'-' + electric['Unnamed: 1'] 
electric.index = electric['Combined']
electric.index=pd.to_datetime(electric.index, format = '%Y-%b')
electric
```

```python
import numpy as np
weather = pd.read_csv('CLMTEMP_HKO_.csv')
#csv from https://data.gov.hk/tc-data/dataset/hk-hko-rss-daily-temperature-info-hko/resource/7f42664a-04ec-4bb4-ac35-d257a554c661
#weather2=weather.groupby(["月/Month"])['數值/Value'].mean()
weather.dropna(inplace=True)
#weather.set_index(['年/Year','月/Month'])
#weather2=weather.groupby(['年/Year', '月/Month'])["數值/Value"].agg(['max','min'])
weather1=weather[weather['年/Year']>1978]
weather2=weather1[['年/Year', '月/Month','日/Day',"數值/Value"]].astype(dtype=np.float32)
weather3=weather2.groupby(['年/Year', '月/Month'])["數值/Value"].agg(['max', 'min', 'mean'])
weather3.plot()
```

```python
# change the index into datetime
weather.head(3)
weather['Combined'] = weather['年/Year'].astype(str) +'-' + weather['月/Month'].astype(str) 
pd.to_datetime(weather['Combined'], format = '%Y-%m')
weather.set_index(weather['Combined'])
weather.index = pd.to_datetime(weather['Combined'], format = '%Y-%m')
weather

```

```python
fig, ax_list = plt.subplots(3,1)
ax_up = ax_list[0]
ax_down = ax_list[2]
ax_middle = ax_list[1]
ax_up.plot(electric.index, electric['All groups'])
ax_middle.plot(weather.index, weather['數值/Value'])
ax_down.plot(electric.index, electric['All groups.1'])
plt.show()
```

```python
#plot starting from 2020 to now

electric3=electric[electric.index.year>2020]
weather4 = weather[weather.index.year>2020]
fig, ax_list = plt.subplots(2,1)
ax_up = ax_list[0]
ax_down = ax_list[1]

ax_up.set_title('electricity usage by all groups')
ax_up.plot(electric3.index, electric3['All groups'])
ax_down.plot(electric3.index, electric3['All groups.1'])
ax_down.set_title('gas usage by all groups')
plt.show()
```
