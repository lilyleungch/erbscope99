```python
import numpy as np
weather = pd.read_csv('CLMTEMP_HKO_.csv')
#csv from https://data.gov.hk/tc-data/dataset/hk-hko-rss-daily-temperature-info-hko/resource/7f42664a-04ec-4bb4-ac35-d257a554c661
#weather2=weather.groupby(["/Month"])['/Value'].mean()
weather.dropna(inplace=True)
#weather.set_index(['/Year','/Month'])
#weather2=weather.groupby(['/Year', '/Month'])["/Value"].agg(['max','min'])
weather1=weather[weather['/Year']>1978]
weather2=weather1[['/Year', '/Month','/Day',"/Value"]].astype(dtype=np.float32)
weather3=weather2.groupby(['/Year', '/Month'])["/Value"].agg(['max', 'min', 'mean'])
weather3.plot()
```
