# Add a column 

```python
df_hsi["Return"] = df_hsi["Adj Close"].pct_change(1)   # percentage change from 1 day before
df_hsi
```

```python
# add 10,20,50 days moving average
for period in [10,20,50]:
    df_hsi[f"MA{period}"] = df_hsi["Adj Close"].rolling(period).mean()
df_hsi.tail(5)
```

# Concat DataFrame

```python
import yfinance as yf
df_gold_jan = yf.download("gold", start="2023-01-01", end="2023-01-31")
df_gold_jan.head(3)
```

```python
df_gold_feb = yf.download("gold", start="2023-02-01", end="2023-02-28")
df_gold_feb.head(3)
```

```python
df_gold = pd.concat([df_gold_jan, df_gold_feb])
df_gold
```

```python
df_gold.shape
```

```python
df_buysell = pd.DataFrame({'BullSell' : np.zeros(37)})    # Create a DF having same rows
df_buysell 
```

```python
df_gold = pd.concat([df_gold.reset_index(), df_buysell], axis=1).set_index('Date')    # concat 2 df with same index
df_gold
```
