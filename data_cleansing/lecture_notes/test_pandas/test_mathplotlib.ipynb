```python
font1 = {
    'family'  : 'serif',
    'color'  : 'r',
    'size' : 20,
    'weight' : 'bold'
}

font2 = {'family' : 'Arial',
         "color" : 'darkred',
         'weight': 'normal',
         'size':16
         }


ypoints = np.array([3,8,1,10])
ypoints2 = np.array([1,4,7,5])
plt.plot(ypoints, 'r', lw = 20)
plt.plot(ypoints2, 'g',lw=20, ms = 50, markerfacecolor = 'navy', marker = 'o')
plt.title('Income vs year', fontdict = font1)
plt.xlabel('year', fontdict = font2)
plt.ylabel('Income', fontdict = font2)
plt.show()
```

```python
#Implicit approach
x= [20,30,40]
y=[100, 90, 70]

plt.plot(x,y, 'o:y')
plt.xlabel ('Age')
plt.ylabel('IQ')
plt.title('My IQ vs age')
plt.show()
```

```python
#explicit approach
fig, ax = plt.subplots()
x= [20,30,40]
y=[100, 90, 70]
ax.plot(x,y, 'o:y')
ax.set_xlabel('Age')
ax.set_ylabel('IQ')
ax.set_title('Graph2')
plt.show()

```

```python
from matplotlib import font_manager
fontP = font_manager.FontProperties(family='Microsoft JhengHei', size=14)
# Microsoft JhengHei is a built-in font
fig = pd.DataFrame({
    '':[1],
    '':[2],
    '':[3],
    '':[4]})
fig.plot()
plt.legend(loc=0, prop=fontP)
plt.title('',
          fontproperties='MingLiu',
          fontsize='xx-large')
plt.grid(True)
plt.axis('tight')
```
