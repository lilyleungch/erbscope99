```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
ticker='GOOG'
yfdata = yf.download(ticker, start="2022-10-01", end="2022-12-01")
yfdata.to_csv(f'{ticker}.csv')
data = pd.read_csv("GOOG.csv")
print(data.head())
```

```python
fig, (ax1, ax2)= plt.subplots(2,1,sharex = True, figsize = (15, 10))
ax1.plot(data.Date, data.Close)
ax2.bar(data.Date, data.Volume, color = 'darkgrey')
ax1.grid(True)
ax1.set_ylabel('Share price')
ax2.set_ylabel('Daily Volume')
ax2.grid(True)
fig.suptitle('Alphabet Inc. GOOG')
plt.xticks(rotation = 45, ha = 'right')
plt.plot()
```
