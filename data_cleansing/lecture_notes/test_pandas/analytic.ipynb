```python
def findslope(x1, x2, y1, y2):
    m= (y2-y1)/(x2-x1)
    return m

def findintercept(x1, x2, y1, y2):
    c = y2-(x2*(y2-y1)/(x2-x1))
    return c
print(findslope(1,2,3,5))
print(findintercept(1,2,3,5))
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

ticker='QQQ'

yfdata = yf.download(ticker, start="2024-08-12", end="2024-10-04")
yfdata.to_csv(f'{ticker}.csv')

data = pd.read_csv("QQQ.csv")
print(data.head())

```
